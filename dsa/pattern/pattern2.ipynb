```python
n=4
for i in range(n):
  for j in range(1,n+1):
    chrac=chr(ord('A')+j-1)
    print(chrac,end="")
  print()
```

```python
n=7
for i in range(1,n+1):
  chrac=chr(ord('A')+i-1)
  for j in range(i):
    print(chrac,end="")
  print()
```
